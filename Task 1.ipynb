in this task we will Predict the percentage of marks of the students based on the number of hours they studied 

```python
# Importing all libraries required 
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
# Reading the data
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df.head()
```

```python
# Hours vs Percentage Scatter plot
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

## Building our model

```python
#dividing the data to x and y
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values 
```

```python
#importing the function needed to split the data into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
#model training 
from sklearn.linear_model import LinearRegression  
regression = LinearRegression()  
regression.fit(X_train, y_train) 
print('----- done training ------')
```

```python
# Plotting regression line
line = regression.coef_*X+regression.intercept_
plt.scatter(X, y)
plt.plot(X,line,c='r');
plt.show()
```

```python
y_pred = regression.predict(X_test) # Predicting the scores
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

## What will be predicted score if a student studies for 9.25 hrs/ day? 

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score =",round(answer[0],2))
```

```python
# model evaluation 
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

**conclusion** : thoughout the data we notice that there are a linear relation with the hours student spend for studing and their grade and we succsisfully manage to build a modle to predict the grades based on the hours spent with Mean Absolute Error of 4.183859899002975
